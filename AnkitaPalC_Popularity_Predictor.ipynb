```python
import pandas as pd
import numpy as np
```

```python
#Pulling data from shared_articles.csv file

shared_articles_df = pd.read_csv('shared_articles.csv')
```

```python
#Considering data with only CONTENT SHARED and CONTENT REMOVED event Type
shared_articles_df = shared_articles_df[(shared_articles_df['eventType'] == 'CONTENT SHARED') | (shared_articles_df['eventType'] =='CONTENT REMOVED')]
shared_articles_df.head(10)
```

```python
#Dropping columns with NaN/blank values
shared_articles_df = shared_articles_df.dropna(axis='columns')
```

```python
shared_articles_df.head(10)
```

```python
#Pulling data from users_interactions.csv file and dropping unnecessary columns with mostly blank values
user_interactions_df = pd.read_csv('users_interactions.csv')
user_interactions_df = user_interactions_df.dropna(axis='columns')
```

```python
user_interactions_df.head(15)
```

```python
print('Number of interactions:', len(user_interactions_df))
```

```python
#Numbers of users making interactions for different content
users_interactions_count_df = user_interactions_df.groupby(['personId', 'contentId']).size().groupby('personId').size()
print('Numbers of users making interactions for different content:', len(users_interactions_count_df))
users_interactions_count_df
```

```python
#Count of users who usually makes content viral
users_with_10_interactions_df = users_interactions_count_df[users_interactions_count_df >= 10].reset_index()[['personId']]
print('Total relevant users with at least 10 interactions:', len(users_with_10_interactions_df))
```

```python
#person ID making interactions
users_interactions_count_df = users_interactions_count_df.reset_index()[['personId']]
users_interactions_count_df
```

```python
#Merging the two files provided with respect to contentId column
interactions_content_df = shared_articles_df.merge(user_interactions_df, 
               how = 'right',
               left_on = 'contentId',
               right_on = 'contentId')
interactions_content_df.head(5)
```

```python
len(interactions_content_df)
```

```python
interactions_content_df['COUNT'] = 0
```

```python
#From the interactions_content_df building dataframe for all events
user_interactions_df_grp = interactions_content_df.groupby(['contentId','eventType_y'])['COUNT'].size().reset_index().sort_values(by = ['COUNT'],ascending = False)
user_interactions_df_grp
```

```python
#eventType as dataframe
eventType_df = user_interactions_df_grp.pivot_table('COUNT', ['contentId'], 'eventType_y').fillna(0)
eventType_df
```

```python
#relationship represented of event types
import seaborn as sns
import matplotlib.pyplot as plt 
sns.pairplot(eventType_df)
```

```python
#co-relation among the event types
plt.figure(figsize=(12,10))
sns.heatmap(eventType_df.corr(),annot=True,fmt=".2f")
```

```python
#calculate virality
def virality(df):
    vir = (1* df['VIEW']) + (4*df['LIKE']) + (10*df['COMMENT CREATED']) +( 25*df['FOLLOW'] )+ (100*df['BOOKMARK'])
    return vir
```

```python
#adding column calculating virality
eventType_df['virality'] = eventType_df.apply (lambda x: virality(x), axis=1)
```

```python
eventType_df.sort_values(by = ['virality'],ascending = False)
```

```python
#split the dataframe into training and testing dataset
from sklearn.model_selection import train_test_split

train, test = train_test_split(eventType_df, test_size=0.2)
```

```python
#selection of featres
train_X, train_Y = train.drop('virality',axis = 1), train['virality']
test_X, test_Y = test.drop('virality',axis = 1), test['virality']
train_X,train_Y
```

```python
#creating dictionary for coefficients and measurement metrices
coef = dict()
metrics = dict()
#Coefficients from ground truth
coef['gt'] = [100,10,25,4,1]
```

```python
#Using Linearregression as the model provided is multilinear
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score,explained_variance_score
#Train model
lr = LinearRegression()
lr.fit(train_X,train_Y)

print('Train Score',lr.score(train_X,train_Y))
#Test
predict_test = lr.predict(test_X)

print('predict test 5 samples',predict_test[:5])
print('Test Score',lr.score(test_X,test_Y))

coef['lr'] = lr.coef_
metrics['lr'] = r2_score(test_Y, predict_test),explained_variance_score(test_Y,predict_test)
```

```python
#Using Lasso regression

from sklearn.linear_model import Lasso
#Train model
clf = Lasso(alpha=0.01)
clf.fit(train_X, train_Y)
print('Train Score',clf.score(train_X, train_Y))
#Test
predict_clf = clf.predict(test_X)
print(clf.coef_)
print('Test Score',clf.score(test_X, test_Y))
coef['lasso'] = clf.coef_
metrics['lasso'] = r2_score(test_Y, predict_clf),explained_variance_score(test_Y,predict_clf)
```

```python
print ("Coefficients of the features for both the models")
for c in coef.items():
    print(c)
```

```python
print ("R-square and explained variance score of models:")
for m in metrics.items():
    print(m)
```

```python
#Predict New article's virality

new_article = [117,5,112,12,1]#BOOKMARK	COMMENT CREATED	FOLLOW	LIKE	VIEW

new_article_df = np.array(new_article).reshape(1,-1)


predict_new_article = lr.predict(new_article_df)
print('Predicted virality with linear regression',predict_new_article)

predict_new_article = clf.predict(new_article_df)
print('Predicted virality with Lasso model',predict_new_article)
```
